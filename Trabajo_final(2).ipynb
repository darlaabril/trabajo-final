## Modificar la variable pobreza

La variable de `pobreza`, en la sumaria, está codificada como integers 1, 2 y 3. Esto corresponde a:  
`1`: hogar pobre extremo  
`2`: hogar pobre no extremo  
`3`: hogar no  pobre  

````python
sumaria["pobreza3"] = sumaria.pobreza.replace({1:"pobre extremo",2:"pobre no extremo",3:'no pobre'})
sumaria
````
